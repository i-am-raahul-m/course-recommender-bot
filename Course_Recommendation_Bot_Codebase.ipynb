# Imports

```python
import os
import torch
from langchain import PromptTemplate
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
```

# Global Constants

```python
root_dir = "/content/drive/MyDrive/IBM_Gen_AI_Project"
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

# Creating Persistent Vector Stores


```python
# If run when the vector database already exists in the drive folder, then does nothing
from langchain.document_loaders import TextLoader, PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

texts_dir = os.path.join(root_dir, "RAG_texts")
persist_path = os.path.join(root_dir, "chroma_db")

if os.path.exists(texts_dir) and not os.path.exists(persist_path):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )
    docs = []

    for filename in os.listdir(texts_dir):
        filepath = os.path.join(texts_dir, filename)

        if filename.endswith(".txt"):
            try:
                loader = TextLoader(filepath)
                text_docs = loader.load()
                docs.extend(text_docs)
                print(f"Loaded {len(text_docs)} text chunks from {filename}")
            except Exception as e:
                print(f"Failed to load TXT file {filename}: {e}")

        elif filename.endswith(".pdf"):
            try:
                loader = PyPDFLoader(filepath)
                pdf_docs = loader.load()
                docs.extend(pdf_docs)
                print(f"Loaded {len(pdf_docs)} PDF chunks from {filename}")
            except Exception as e:
                print(f"Failed to load PDF file {filename}: {e}")

    if not docs:
        print(f"No documents found in {texts_dir}.")

    else:
      docs = splitter.split_documents(docs)

      os.makedirs(persist_path, exist_ok=True)

      db = Chroma.from_documents(
          documents=docs,
          embedding=embedding,
          persist_directory=persist_path
      )
      db.persist()

      print(f"DB content: {db._collection.count()} vectors saved.")
      print(f"Saved vector DB to: {persist_path}")
```

# Vector Store

```python
texts_dir = os.path.join(root_dir, "RAG_texts")
persist_path = os.path.join(root_dir, "chroma_db")

store = None
if os.path.isdir(persist_path):
    store = Chroma(
        persist_directory=persist_path,
        embedding_function=embedding
    )

print(f"RAG store: {store._collection.count()} vectors")
```

# Keyword Extractor: Flan-t5-large

```python
model_name = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

hf_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=20,
)

keyword_extractor_llm = HuggingFacePipeline(pipeline=hf_pipeline)
```

# Few Shot Prompt for Keyword Extraction

```python
keyword_prompt = PromptTemplate.from_template("""
Extract keywords related to mental health from the given user prompt.
Follow the given examples:

user prompt: I want to learn about the concept of virtualization and containerization. Which course is recommendable?
keywords: virtualization, containerization

user prompt: I'm curious about how load balancers distribute traffic across multiple servers. Can you explain?
keywords: load balancers, traffic distribution

user prompt: How does memory segmentation work when switching between processes?
keywords: memory segmentation, process switching

user prompt: What are the differences between symmetric and asymmetric encryption in securing communications?
keywords: symmetric encryption, asymmetric encryption

user prompt: Is there a simple guide to understanding the fetch-decode-execute cycle in modern processors?
keywords: fetch-decode-execute cycle, processors

user prompt: I'm trying to wrap my head around how IP addressing and subnetting work. Any tips?
keywords: IP addressing, subnetting

user prompt: Could you explain the benefits of using message queues in distributed systems?
keywords: message queues, distributed systems

user prompt: What's the difference between kernel-mode and user-mode in an OS?
keywords: kernel-mode, user-mode

user prompt: I want to study how consistency and availability trade-offs are handled in scalable systems.
keywords: consistency, availability, scalable systems

user prompt: How do instruction pipelines help improve performance in CPUs?
keywords: instruction pipelines, CPU performance

user prompt: {query}
keywords:
""")

keyword_chain = keyword_prompt | keyword_extractor_llm
```

# Recommendation LLM: Mistral-7B-Instruct-v0.2

```python
# Loading base model from huggingface

base_model_id = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    quantization_config=bnb_config,
    torch_dtype=torch.float16,
    use_auth_token = True,
    device_map="auto",
    max_length = 4096
)

tokenizer = AutoTokenizer.from_pretrained(base_model_id)
```

```python
# Creating a chain

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length = 4096,
    pad_token_id=tokenizer.eos_token_id
)

recommendation_llm = HuggingFacePipeline(pipeline=pipe)
```

# Structured System Prompt for Final Response

```python
final_prompt = PromptTemplate.from_template("""
INSTRUCTION:
You are an expert Course Recommendation Bot.
Provide an informative reply suggesting the best learning paths involving courses and certifications based on user requirements, interests and background.
If the user specifies a number of courses to recommend, strictly adhere to that number of most relevant suggestions.
Use the extracted KEYWORDS and curriculum details of AVAILABLE COURSES AND CERTIFICATIONS to answer the USER QUERY.

USER QUERY:
{query}

KEYWORDS:
{keywords}

AVAILABLE COURSES AND CERTIFICATIONS:
{context}

RESPONSE:
""")

generation_chain = final_prompt | recommendation_llm
```

# RAG Pipeline

```python
def rag_pipeline(user_query: str, keyword_str):

    keywords = [word.strip() for word in keyword_str.split(',')]

    def count_keyword_matches(text, keywords):
        return sum(1 for kw in keywords if kw.lower() in text.lower())

    def retrieve_for_store(store):
        if store is not None:
            results = store.similarity_search(user_query, k=10)
            reranked_results = sorted(
                results,
                key=lambda chunk: count_keyword_matches(chunk.page_content, keywords),
                reverse=True
            )
            return reranked_results
        else:
          return []


    retrieved_chunks = retrieve_for_store(store)

    combined_context = ""
    for chunk in retrieved_chunks:
        combined_context += f"{chunk}:\n"

    # print(combined_context)
    return combined_context
```

# RAG Testing

```python
query = "I want to learn about cloud architectures and deployment model."
keywords = "cloud, architecture, deployment, model"
print("\nOutput:\n", rag_pipeline(query, keywords))
```

# FINAL RESPONSE RESULT

```python
def wrap_around(text, max_chars=140):
    result = []
    lines = text.split("\n")
    for line in lines:
        while len(line) >= max_chars:
            ind = line[:max_chars].rfind(" ")
            if ind == -1:
                ind = max_chars
            result.append(line[:ind])
            line = line[ind+1:]
        result.append(line)
    return "\n".join(result)
```

```python
user_query = "I am computer network administrator, looking to enhance my resume. What certifications for computer networks are available?"

# Keyword Extraction
extracted_keywords = keyword_chain.invoke(user_query)

# Context Retrieval (RAG)
retrieved_context = rag_pipeline(user_query, extracted_keywords)

# LLM Final Response
response = generation_chain.invoke({"query": user_query, "keywords": extracted_keywords,"context": retrieved_context})
if "RESPONSE:" in response:
    filtered_response = response.split("RESPONSE:")[-1].strip()
else:
    filtered_response = response.strip()

# Display the final response

print("QUERY:", user_query)
print("RESPONSE:")
print(wrap_around(filtered_response))
```
